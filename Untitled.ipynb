```python
import geopandas as gpd
import matplotlib.pyplot as plt
import requests

# Define the search query for pharmacies in Youssoufia
search_query = "marché"
city_name = "Youssoufia"
url = f"https://nominatim.openstreetmap.org/search?q={search_query},{city_name}&format=json"

# Fetch data from OpenStreetMap API (Nominatim)
response = requests.get(url)
data = response.json()

# Extract coordinates and pharmacy names
pharmacies = []
for place in data:
    pharmacy_name = place.get("display_name", "Unknown Pharmacy")
    lon, lat = place["lon"], place["lat"]
    pharmacies.append({"name": pharmacy_name, "lon": lon, "lat": lat})

# Create a GeoDataFrame with the geometry column
gdf = gpd.GeoDataFrame(pharmacies, geometry=gpd.points_from_xy([float(pharmacy["lon"]) for pharmacy in pharmacies], 
                                                         [float(pharmacy["lat"]) for pharmacy in pharmacies]),
                       crs="EPSG:4326")


gdf
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import requests

# Define the search query for pharmacies in Youssoufia
search_query = "pharmacie"
city_name = "Youssoufia"
url = f"https://nominatim.openstreetmap.org/search?q={search_query},{city_name}&format=json"

# Fetch data from OpenStreetMap API (Nominatim)
response = requests.get(url)
data = response.json()

# Extract coordinates and pharmacy names
pharmacies = []
for place in data:
    pharmacy_name = place.get("display_name", "Unknown Pharmacy")
    lon, lat = place["lon"], place["lat"]
    pharmacies.append({"name": pharmacy_name, "lon": lon, "lat": lat})

# Create a GeoDataFrame with the geometry column
gdf = gpd.GeoDataFrame(pharmacies, geometry=gpd.points_from_xy([float(pharmacy["lon"]) for pharmacy in pharmacies], 
                                                         [float(pharmacy["lat"]) for pharmacy in pharmacies]),
                       crs="EPSG:4326")

# Create a map to display the pharmacies
fig, ax = plt.subplots(figsize=(10,10))
youssoufia = gpd.read_file('youssoufia.geojson')
youssoufia.plot(ax=ax, alpha=0.4)
gdf.plot(ax=ax, markersize=10, color='red', marker='o', label='Pharmacies')
ax.set_title('Pharmacies in Youssoufia')
ax.legend()
plt.show()
```

```python
import overpy

# Define the query to extract restaurants within a 2 km radius of Youssoufia
query = """
    [out:json];
    (
        node["amenity"="restaurant"](around:2000,33.2278,-8.5104);
        way["amenity"="restaurant"](around:2000,33.2278,-8.5104);
        relation["amenity"="restaurant"](around:2000,33.2278,-8.5104);
    );
    out center;
"""

# Define the Overpass API endpoint
api = overpy.Overpass()

# Execute the query and extract the coordinates of the restaurants
result = api.query(query)
for restaurant in result.nodes + result.ways + result.relations:
    print(restaurant.center_lat, restaurant.center_lon)
```

```python
import folium
import overpy

# Create an Overpass API object
api = overpy.Overpass()

# Define the Overpass query to extract nodes with the tag amenity=cafe
query = """
[out:xml][timeout:25];
area["name"="Youssoufia"]->.a;
node(area.a)["amenity"="gare"];
out body;
>;
out skel qt;
"""

# Execute the query and get the results
result = api.query(query)

# Create a folium map centered on the city of Youssoufia
map = folium.Map(location=[32.249, -8.535], zoom_start=14)

# Create a marker for each cafe in the results
for node in result.nodes:
    lat = node.lat
    lon = node.lon
    name = node.tags.get("name", "Unknown Cafe")
    marker = folium.Marker([lat, lon], popup=name)
    marker.add_to(map)

# Display the map
map.save("cafes.html")
```

```python
import requests
import pandas as pd

# Requête Overpass API pour récupérer les nœuds ayant le tag amenity=cafe dans la ville de Youssoufia
overpass_url = "https://overpass-api.de/api/interpreter"
overpass_query = """
[out:json];
area["name"="Youssoufia"]->.a;
(
  node["amenity"="cafe"](area.a);
);
out center;
"""
response = requests.get(overpass_url, params={'data': overpass_query})
data = response.json()

# Création d'une liste de dictionnaires contenant les informations de chaque nœud
cafes = []
for node in data['elements']:
    if 'lat' in node and 'lon' in node:
        lat = node['lat']
        lon = node['lon']
        name = node.get('tags', {}).get('name', 'Unknown Cafe')
        cafes.append({'name': name, 'lat': lat, 'lon': lon})

# Création d'un dataframe à partir de la liste de dictionnaires
df = pd.json_normalize(cafes)

# Affichage du dataframe
print(df.head())
```

```python
import json
import folium

# Lecture des données JSON depuis un fichier
with open('information.json') as f:
    data = json.load(f)

# Création de la carte avec folium
m = folium.Map(location=[32.247416547047806, -8.534593067824346], zoom_start=14)

# Ajout des marqueurs pour chaque élément
for feature in data['features']:
    name = feature.get('properties', {}).get('name', 'Unknown')
    coords = feature['geometry']['coordinates']
    color = feature.get('properties', {}).get('marker-color', 'green')
    folium.Marker(location=[coords[1], coords[0]], tooltip=name, icon=folium.Icon(color=color)).add_to(m)

# Affichage de la carte
m
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Lecture des données JSON depuis un fichier
with open('information.json') as f:
    data = json.load(f)

# Extraction des coordonnées et des noms de chaque élément
coords = np.array([feature['geometry']['coordinates'] for feature in data['features']])
names = np.array([feature['properties']['name'] for feature in data['features']])

# Création du graphique avec Matplotlib
fig, ax = plt.subplots(figsize=(8, 8))

# Tracé des points sur le graphique
for name in np.unique(names):
    mask = names == name
    ax.scatter(coords[mask, 0], coords[mask, 1], s=50, alpha=0.8, label=name)

# Configuration du graphique
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Points d\'intérêt')
ax.legend()

# Affichage du graphique
plt.show()
```
